```python
datasets = {
    "customers": customers,
    "geolocation": geolocation,
    "items": items,
    "order_payments": order_payments,
    "orders": orders,
    "order_reviews": order_reviews,
    "products": products,
    "sellers": sellers,
    "category_name_translation": category_name_translation
}

for name, df in datasets.items():
    print(f"\n===== {name.upper()} =====")
    print(df.head())
    print("\nCount:")
    print(df.count())
    print("\nDtypes:")
    print(df.dtypes)
```

```python
def check_pk(df, column_name):
    
    # Nếu là 1 cột (string)
    if isinstance(column_name, str):
        is_unique = df[column_name].is_unique
        has_null = df[column_name].isnull().any()
    
    # Nếu là nhiều cột (list)
    else:
        is_unique = not df.duplicated(subset=column_name).any()
        has_null = df[column_name].isnull().any().any()
    
    print(f"Column(s): {column_name}")
    print(f"- Unique: {is_unique}")
    print(f"- Has Null: {has_null}")
    
    if is_unique and not has_null:
        print("=> Kết luận: Đủ điều kiện làm Primary Key.")
    else:
        print("=> Kết luận: CẢNH BÁO! Không thể dùng làm PK trực tiếp.")

# Thử với bảng order_items
check_pk(customers,"customer_id")
check_pk(geolocation,"geolocation_zip_code_prefix")
check_pk(orders, "order_id")
check_pk(items,["order_id","order_item_id"])
check_pk(order_payments,["order_id","payment_sequential"])
check_pk(sellers,"seller_id")
check_pk(products,"product_id")
check_pk(category_name_translation,"product_category_name")
check_pk(order_reviews,["review_id","order_id"])
```
